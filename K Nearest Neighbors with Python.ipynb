## Import Libraries



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
#import settingan untuk ukuran gambar juga bisa dilakukan di awal
```

```python
df = pd.read_csv("Classified Data",index_col=0)
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#untuk men-standardisasi data agar standar deviasi nya menjadi 1
#yang distandardisasi adalah fitur
scaler = StandardScaler()
```

```python
scaler.fit(df.drop('TARGET CLASS',axis=1))
```

```python
x_call = df.columns[:-1]
```

```python
x = df[x_call]
y = df['TARGET CLASS']
```

```python
scaled_features = scaler.transform(x)
#bisa juga menggunakan scaled_features = scaler.fit_transform(x)
```

```python
df_feat = pd.DataFrame(scaled_features,columns=x_call)
df_feat.head()
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_feat,df['TARGET CLASS'],
                                                    test_size=0.30,random_state=207)
```

```python
x_train
```

## Using KNN

```python
from sklearn.neighbors import KNeighborsClassifier,RadiusNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=1,p=2)
```

```python
model.fit(x_train,y_train)
```

# Cek training error

```python
pred_train=model.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_train,pred_train)
```

```python
pred = model.predict(x_test)
```

## Predictions and Evaluations

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```
