# DATA PREPROCESSING

```python
df.columns
```

```python
y = df['CREDIT_SCORE']
X = df[['DURATION_IN_MONTH',
       'CREDIT_AMOUNT',
       'INSTALLMENT_RATE_IN_PERCENTAGE_OF_DISPOSABLE_INCOME',
       'AGE_IN_YEARS',
       'NUMBER_OF_EXISTING_CREDITS_AT_THIS_BANK',
       'NUMBER_OF_PEOPLE_BEING_LIABLE_TO_PROVIDE_MAINTENANCE_FOR']]
X_trans = df[['STATUS_OF_EXISTING_CHECKING_ACCOUNT',
       'CREDIT_HISTORY','SAVINGS_ACCOUNT/BONDS',
       'PRESENT_EMPLOYMENT_SINCE',
       'OTHER_DEBTORS/GUARANTORS',
       'PROPERTY',
       'OTHER_INSTALLMENT_PLANS', 'HOUSING',
       'JOB','TELEPHONE',
       'FOREIGN_WORKER']]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
X_trans_le = X_trans.apply(LabelEncoder().fit_transform)
```

```python
y_trans_le = LabelEncoder().fit_transform(y)
```

```python
y_trans_le
```

```python
X_trans_le
```

```python
X_trans_col=X_trans.columns
```

```python
df = df.drop(X_trans,axis=1)
```

```python
X = pd.concat([X,X_trans_le],axis = 1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y_trans_le,test_size=0.3,random_state=217,stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler
standard = StandardScaler()
X_train = standard.fit_transform(X_train)
X_test = standard.transform(X_test)
```

```python
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
knn = KNeighborsClassifier()
score = cross_val_score(knn,X_train,y_train,cv=5,scoring='accuracy')
print(score.mean(),score.std())
```

```python
score
```

```python
knn.fit(X_train,y_train)
pred = knn.predict(X_test)
print('Accuracy:',metrics.accuracy_score(y_test,pred))
```

```python
pred_prob = knn.predict_proba(X_test)[:,1]
print('ROC_AUC:',metrics.roc_auc_score(y_test,pred_prob))
```

```python
logreg = LogisticRegression(solver='lbfgs',multi_class='multinomial')
score = cross_val_score(logreg,X_train,y_train,cv=5,scoring='accuracy')
print(score.mean(),score.std())
```

```python
logreg.fit(X_train,y_train)
pred = logreg.predict(X_test)
print('Accuracy:',metrics.accuracy_score(y_test,pred))
```

```python
pred_prob = logreg.predict_proba(X_test)[:,1]
print('ROC_AUC:',metrics.roc_auc_score(y_test,pred_prob))
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
clf = LogisticRegression(solver='lbfgs',multi_class='multinomial')
```

```python
ada_clf = AdaBoostClassifier(clf, random_state=217)
```

```python
score = cross_val_score(ada_clf,X_train,y_train,cv=5,scoring='accuracy')
print(score.mean(),score.std())
```

```python
ada_clf.fit(X_train,y_train)
pred = ada_clf.predict(X_test)
```

```python
metrics.accuracy_score(y_test,pred)
```

```python
pred_prob = ada_clf.predict_proba(X_test)[:,1]
print('ROC_AUC:',metrics.roc_auc_score(y_test,pred_prob))

fpr_rf, tpr_rf, _ = metrics.roc_curve(y_test, pred_prob)

plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr_rf, tpr_rf, label='RF')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC curve')
plt.show()
```
